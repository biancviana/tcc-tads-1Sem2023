# Análise das variáveis/indicadores por regressão linear

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 40)
#pd.set_option('display.max_rows', 50)
```

```python
# carrega dataset original
df = pd.read_csv('../../datasets/bullying.csv', sep=';')

df
```

```python
df.info()
```

```python
# mostre todas as variáveis
df.columns
```

```python
# nossa variável target/destino será a média das 3 notas dos alunos
df['nota-media'] = df[['Nota-exatas', 'Nota-humanas', 'Nota-biológicas']].mean(axis=1)
df['nota-media']
```

```python
# cria uma coluna com a quantide total de bullyings sofridos
# df['Fisico'] + df['Social'] + ...
df['total-de-bullyings'] = df[['Fisico', 'Social', 'Verbal', 'Material', 'Virtual']].sum(axis=1)
df
```

```python
# Sofreu bullying?
# se o aluno sofreu ao menos bullying uma vez, vamos considerar que ele/a sofre bullying
df['sofreu-bullying'] = df['total-de-bullyings'] >= 1
df
```

```python
variaveis_categoricas = ['Demora dormir', 'Sonhos ruins', 'Qualidade do sono', 'sofreu-bullying']
df[variaveis_categoricas]
```

```python
# extraia apenas as colunas desejadas na análise: variaveis_numericas, variaveis_categoricas
variaveis_de_interesse = variaveis_categoricas + ['nota-media']

df_analise = df[variaveis_de_interesse].copy()
df_analise
```

```python
# realize o one-hot encoding em cima das variáveis categóricas
df_analise = pd.get_dummies(df_analise, columns=variaveis_categoricas)
df_analise
```

```python
# dividindo o dataset entre treino e teste
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df_analise, train_size=0.8, random_state=42)
```

```python
df_train
```

```python
df_test
```

```python
# montando a matriz de features de treinamento e o array de labels para predição
y_train = df_train['nota-media']
y_train
```

```python
X_train = df_train.drop(columns='nota-media')
X_train
```

```python
# x0 == Fisico
# x1 == Social
# ...
feature_names = X_train.columns.to_list()
feature_names
```

```python
# teste
X_test = df_test.drop(columns='nota-media')
y_test = df_test['nota-media']
```

```python
# treinando um random forest
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor()

regressor.fit(X_train, y_train)
```

```python
# predição do conjunto de treinamento
y_train_pred = regressor.predict(X_train)
```

```python
# coeficiente de determinação
# quanto mais próximo de 1, significa que as variáveis de estudo explicam melhor a variável de saída
from sklearn.metrics import r2_score

r2_train = r2_score(y_train, y_train_pred)
print(f'Coeficiente de Determinação para o Treinamento: {r2_train}')
```

```python
# medindo o erro de predição para o treinamento (que são amostras já vistas durante o treinamento)
from sklearn.metrics import mean_absolute_error

mae_train = mean_absolute_error(y_train, y_train_pred)
print(f'Mean Absolute Error para o Treinamento: {mae_train}')
```

```python
# predição do conjunto de treinamento
y_test_pred = regressor.predict(X_test)
```

```python
# medindo o erro de predição para o teste
from sklearn.metrics import mean_absolute_error

mae_test = mean_absolute_error(y_test, y_test_pred)
print(f'Mean Absolute Error para o Teste: {mae_test}')
```
